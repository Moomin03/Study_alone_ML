Îã§Ï§ë Î∂ÑÎ•òÏóêÏÑúÎäî ÌÅ¨Î°úÏä§ÏóîÌä∏Î°úÌîº ÏÜêÏã§Ìï®ÏàòÎ•º ÏÇ¨Ïö©ÌïòÏ£†.

```python
import pandas as pd
fish = pd.read_csv('https://bit.ly/fish_csv_data')
fish.head()
```

```python
fish_input = fish[['Weight', 'Length', 'Diagonal', 'Height', 'Width']].to_numpy()
fish_target = fish['Species'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(fish_input,
                                                                      fish_target,
                                                                      random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
train_scaled = ss.fit_transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
# ÌôïÎ•†Ï†Å Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ïÏùÑ ÏßÄÏõêÌïòÎäî ÎùºÏù¥Î∏åÎü¨Î¶¨Îäî SGDClassifierÏûÖÎãàÎã§.
'''
loss : {'hinge', 'log_loss', 'modified_huber', 'squared_hinge',
        'perceptron', 'squared_error', 'huber', 'epsilon_insensitive',
        'squared_epsilon_insensitive'}, default='hinge'
        - Ï†ÅÏö© Í∞ÄÎä•Ìïú ÏÜêÏã§Ìï®ÏàòÏûÖÎãàÎã§.
        Î∂ÑÎ•ò Î¨∏Ï†ú: hinge, log_loss, perceptron, squared_hinge Îì±ÏùÑ ÏÇ¨Ïö©Ìï©ÎãàÎã§.
        ÌöåÍ∑Ä Î¨∏Ï†ú: squared_error, huber, epsilon_insensitive Îì±ÏùÑ Í≥†Î†§Ìï©ÎãàÎã§.
        Ïù¥ÏÉÅÏπòÏóê ÎØºÍ∞êÌïú Í≤ΩÏö∞: huber, modified_huberÎ•º ÏÇ¨Ïö©Ìï† Ïàò ÏûàÏäµÎãàÎã§.

'''
from sklearn.linear_model import SGDClassifier
sc = SGDClassifier(loss='log_loss', max_iter=10, random_state=42)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

```python
# ÌôïÎ•†Ï†Å Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ïÏùÄ Ï†êÏßÑÏ†Å ÌïôÏäµÏù¥ Í∞ÄÎä•ÌïòÎãà, Ïù¥Ïñ¥ÏÑú Ï†ÅÏñ¥Î≥ºÍπåÏöî?
```

```python
sc.partial_fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```

```python
# partial_fitÏùÑ ÏÇ¨Ïö©ÌïòÎ©¥ 1ÏóêÌè¨ÌÅ¨Ïî© ÏßÑÌñâÌï† Ïàò ÏûàÏäµÎãàÎã§.
# ÏóêÌè¨ÌÅ¨Ïùò ÏàòÍ∞Ä Ï†ÅÏúºÎ©¥ ÌõàÎ†® ÏÑ∏Ìä∏Î•º Îçú ÌïôÏäµÌï©ÎãàÎã§. ÌïòÏßÄÎßå ÏóêÌè¨ÌÅ¨ ÏàòÍ∞Ä Ï∂©Î∂ÑÌûà ÎßéÏúºÎ©¥ Ïñ¥ÎñªÍ≤å Îê†ÍπåÏöî? Í≥ºÏ†ÅÌï©Ïù¥ Î∞úÏÉùÌï©ÎãàÎã§.
# classes Îß§Í∞úÎ≥ÄÏàòÎ•º Î™ÖÏãúÌï¥Ïïº ÌïòÎäî Ïù¥Ïú†Îäî, Î™®Îç∏Ïù¥ Ï≤òÏùå ÌïôÏäµÏùÑ ÏãúÏûëÌï† Îïå Îç∞Ïù¥ÌÑ∞ÏÖãÏóê Ï°¥Ïû¨ÌïòÎäî Î™®Îì† ÌÅ¥ÎûòÏä§ Î†àÏù¥Î∏îÏùÑ ÎØ∏Î¶¨ ÏïåÍ≥† ÏûàÏñ¥Ïïº ÌïòÍ∏∞ ÎïåÎ¨∏ÏûÖÎãàÎã§. 
# Ïù¥Îäî ÌÅ¥ÎûòÏä§Ïùò ÏàòÎÇò Î≤îÏúÑÎ•º ÏïåÏßÄ Î™ªÌïòÎäî ÏÉÅÌô©ÏóêÏÑú ÏòàÏ∏° Î∂àÍ∞ÄÎä•Ìïú Î†àÏù¥Î∏îÏù¥ Îì±Ïû•Ìï† Í≤ΩÏö∞Î•º Î∞©ÏßÄÌïòÍ∏∞ ÏúÑÌï¥ ÌïÑÏöîÌï©ÎãàÎã§.
```

```python
sc = SGDClassifier(loss='log_loss', random_state=42, alpha=0.0001, penalty='l2')
train_score = []
test_score = []
classes = np.unique(train_target)
for _ in range(0, 300):
    sc.partial_fit(train_scaled, train_target, classes=classes)
    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
plt.plot(train_score)
plt.plot(test_score)
```

```python
sc = SGDClassifier(loss='hinge', random_state=42)
train_score = []
test_score = []
classes = np.unique(train_target)
for _ in range(0, 100):
    sc.partial_fit(train_scaled, train_target, classes=classes)
    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
plt.plot(train_score)
plt.plot(test_score)
```

```python
sc = SGDClassifier(loss='perceptron', random_state=42)
train_score = []
test_score = []
classes = np.unique(train_target)
for _ in range(0, 100):
    sc.partial_fit(train_scaled, train_target, classes=classes)
    train_score.append(sc.score(train_scaled, train_target))
    test_score.append(sc.score(test_scaled, test_target))
plt.plot(train_score)
plt.plot(test_score)
```

```python
# Îã§Ï§ë Î∂ÑÎ•òÏùºÎïå ÏÜêÏã§Ìï®ÏàòÎäî log_lossÍ∞Ä Ï†ÅÏ†àÌï©ÎãàÎã§.
# 100ÏùºÎïå Í∞íÎ∂ÄÌÑ∞Îäî ÌÖåÏä§Ìä∏ÏÖã Ï†êÏàòÍ∞Ä ÏùºÏ†ïÌïòÍ≥†, ÌõàÎ†®Îç∞Ïù¥ÌÑ∞Îßå Î≥ÄÌôîÌïòÎÑ§Ïöî(Í≥ºÏ†ÅÌï©)
# Îî∞ÎùºÏÑú max_iterÎ•º 100ÏúºÎ°ú ÎëêÍ≤†ÏäµÎãàÎã§.
# tolÏùò Í∞íÏùÄ Ìñ•ÏÉÅÎê† ÏµúÏÜüÍ∞íÏùÑ ÏßÄÏ†ïÌïòÎäîÎç∞, 0.001Ïù¥ÎùºÎèÑ Ìñ•ÏÉÅÎêòÏßÄ ÏïäÏúºÎ©¥ Î©àÏ∂îÍ∏∞ ÎïåÎ¨∏Ïóê NoneÏúºÎ°ú ÏßÄÏ†ïÌï†Í≤åÏöî.
sc = SGDClassifier(loss='log_loss', max_iter=100, random_state=42, tol=None)
sc.fit(train_scaled, train_target)
print(sc.score(train_scaled, train_target))
print(sc.score(test_scaled, test_target))
```
