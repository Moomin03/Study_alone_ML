# 🐱 앙상블 모델
---
    하나의 모델로 최적의 성능을 내지 못할 때, 우리는 여러개의 모델을 하나로 합칩니다.
    그러한 모델을 우리는 앙상블 모델이라고 부릅니다.
    대표적인 앙상블 방법으로는 3가지가 있는데, 배깅, 부스팅, 랜덤 포레스트가 있습니다.
    제일 많이 사용되는게 랜덤 포레스트, 부스팅이 있겠네요.
---
    이런 방법을 알기 저에 먼저 부트스트랩에 대해서 알아보죠.
    부트스트랩은 데이터 샘플을 랜덤하게 선택하여 여러 개의 서브 샘플을 생성하는건데, 원본 데이터셋에서 샘플을 중복해서 선택하는 특징이 있습니다. 한마디로 훈련시킬 데이터를 고를 때, 데이터를 뽑고, 다시 넣는다는거죠. 
    부트스트랩의 장점으로는 오차감소 : 여러 모델의 예측을 결합함으로써 예측 오차를 줄일 수 있습니다.
    부트스트랩에서는 중복 샘플링이 발생하기 때문에 각 서브샘플이 원본 데이터셋의 특정 샘플을 여러 번 포함할 수 있습니다. 이로 인해 과적합(overfitting)이 발생할 가능성이 있지만, 배깅과 같은 앙상블 기법에서는 이를 효과적으로 관리할 수 있는 몇 가지 이유가 존재합니다.
    1. 다양한 모델 생성 : 부트스트랩을 통해 생성된 각 서브샘플은 서로 다르기 때문에, 각 모델이 다른 데이터 분포에서 학습하게 됩니다. 즉, 동일한 데이터를 반복해서 학습하더라도, 모델의 구조가 다르면 과적합의 영향을 줄일 수 있습니다.
    2. 앙상블 효과 : 여러 개의 모델을 결합함으로써, 개별 모델의 과적합이 평균화되거나 상쇄되는 효과가 있습니다. 이는 예측의 불확실성을 줄여주고, 결과적으로 성능을 향상시키죠.
    3. 정규화 효과 : 배깅은 일반적으로 모델의 분산을 줄이는 효과가 있습니다.. 즉, 모델의 다양성이 높아지면서 과적합의 영향을 줄일 수 있습니다.
    4. 훈련 데이터의 전체적 특성 유지 :부트스트랩을 통해 생성된 각 서브샘플은 원본 데이터셋의 일부 특성을 반영하고 있기 때문에, 전체 데이터셋의 정보를 여전히 활용하게 됩니다. 따라서 데이터의 다양한 패턴을 포착할 수 있습니다.
--- 
    다음으로는 배깅, 부스팅, 랜덤 포레스트에 대해서 설명할게요.
    먼저 그림을 보시죠 (출처 : Julie의 Tech 블로그)

    그림의 차이를 알겠나요?
    배깅은 여러 폴드에 대한 결과값을 판별해서 투표를 하는거죠. 몇번 클래스가 더 많은지요.
    하지만 부스팅은 깊이가 얕은 여러개의 모델을 합쳐서 결과값을 개선해나간다는거죠.
    랜덤 포레스트는 배깅에 랜덤성과 부트스트랩 샘플링을 더한거랍니다.

```python
# 우선 훈련셋과 테스트셋 분류를 하죠.
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
wine = pd.read_csv('https://bit.ly/wine_csv_data')
data = wine[['alcohol', 'sugar', 'pH']].to_numpy()
target = wine['class'].to_numpy()
train_input, test_input, train_target, test_target = train_test_split(data,
                                                                      target,
                                                                      test_size=0.2,
                                                                      random_state=42)
```

```python
# 이전에 학습했던 결정트리
from sklearn.tree import DecisionTreeClassifier
# 교차검증 라이브러리
from sklearn.model_selection import cross_validate
# 결정트리 불러오기
dtc = DecisionTreeClassifier()
# 교차검증으로 각 점수 파악하기
scores = cross_validate(dtc,
                       train_input,
                       train_target,
                       return_train_score=True,
                       n_jobs=-1)
train_mean = np.mean(scores['train_score'])
test_mean = np.mean(scores['test_score'])
print("train score = {}\ntest_score = {}".format(train_mean, test_mean))
dtc.fit(train_input, train_target)
print(dtc.feature_importances_)
```

### 🐥 여기서 궁금증!
1. cross_validate(교차검증)이랑 GridSearchCV(그리드 서치)는 뭐가 다르나요?
2. 교차검증에서 사용되는 폴드 방법은 어떻게 되나요? 그리고 각 교차 검증 과정도 실제로 보여주세요!
---
    A-1) 교차 검증과 그리드 서치는 목적이 다릅니다. 
    교차 검증은 테스트 데이터를 사용하지 않고, 훈련세트에서 훈련세트와 검증 데이터로 폴드 방법을 사용해 나눈 후, 각각에 대하여 훈련세트에 대한 검증데이터의 "점수"를 제공합니다. 그에 반해 그리드 서치는 해당 모델의 최적의 파라미터를 찾기 위한 방법입니다. 결정 트리에 대한 최적의 파라미터를 찾기 위해 이 값, 저 값 넣어보는 것을 말하죠.


    A-2) 하나씩 차례대로 알려드릴게요.
    1. K-Fold
    - 데이터를 K의 폴드(세트)로 나누고, 각 폴드를 한 번씩 검증 세트로 사용하여 K번의 훈련과 검증을 수행하는 방식이다.
    2. Stratified K-Fold
    - K-Fold의 변형으로 클래스의 배율을 유지하면서 데이터를 K개의 폴드로 나누는 방법이다.
    sklear.model_selection에서 train_test_split 함수 기억나시나요? 훈련 세트와 테스트 세트에 대한 클래스 비율이 이상할때, stratify를 사용했는데 그 stratify 맞습니다.
    3. Leave-One-Out (LOO)
    - 데이터셋의 각 샘플을 검증 세트로 사용하고, 나머지 모든 샘플을 훈련 세트로 사용하는 방법입니다.
    K-Fold와의 차이는 K-Fold는 데이터를 K개의 폴드로 나누지만, LOO는 데이터셋의 샘플 수 N에 따라 N개의 폴드가 있어습니다. 즉, LOOCV는 데이터셋의 각 샘플이 별도의 폴드가 됩니다.

```python
from sklearn.model_selection import KFold, StratifiedKFold, LeaveOneOut
# n_splits는 몇개의 폴드를 나눌지를 결정합니다.
kfold = KFold(n_splits=3)
# kfold.split(train_input)은 각 폴드에서 나눈 훈련세트와 검증세트의 인덱스를 반환합니다.
for idx, (train_index, test_index) in enumerate(kfold.split(train_input)):
    print('{}번째 훈련 폴드분류입니다. : {}'.format(idx+1, train_index))
    print('{}번째 검증 폴드분류입니다. : {}'.format(idx+1, test_index))
```

```python
skf = StratifiedKFold(n_splits=3)
for idx, (train_index, test_index) in enumerate(skf.split(train_input, train_target)):
    print('{}번째 훈련 폴드분류입니다. : {}'.format(idx+1, train_index))
    print('{}번째 훈련 폴드분류입니다. : {}'.format(idx+1, test_index))
```

```python
loo = LeaveOneOut()
for idx, (train_index, test_index) in enumerate(loo.split(train_input)):
    print('{}번째 훈련 폴드분류입니다. : {}'.format(idx+1, train_index))
    print('{}번째 훈련 폴드분류입니다. : {}'.format(idx+1, test_index))
```

```python
# 랜덤포레스트 분류
from sklearn.ensemble import RandomForestClassifier
# 기본적으로 n_estimators는 100입니다. n_estimators는 결정 트리의 수를 의미합니다.
# n_jobs는 사용할 cpu 코어수로 -1은 전부 사용한다는 의미입니다.
rf = RandomForestClassifier(n_estimators=50,
                           n_jobs=-1,
                           random_state=42)
# return_train_score는 훈련세트와 검증세트의 점수를 가져온다는 의미입니다.
scores = cross_validate(rf, 
                        train_input, 
                        train_target,
                        return_train_score=True, 
                        n_jobs=-1)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
# 해당 결과값을 보니 과적합 되었군요.
rf.fit(train_input, train_target)
# 각 특성의 중요성을 볼까요?
print(rf.feature_importances_)
# 당도의 중요성이 결정 트리에 비해서 줄어들었군요.
```

```python
# 아까 부트 스트랩 샘플을 사용할 때, 중복을 사용한다고 했죠? 
# 그렇다면 이때 단 한번도 뽑히지 않는 데이터 값들도 있겠군요.
# 그 값들을 OOB 샘플이라고 합니다. 이 남는 샘플을 사용해서 결정 트리를 평가할 수도 있습니다.
rf = RandomForestClassifier(oob_score=True, 
                            n_jobs=-1, 
                            random_state=42)
rf.fit(train_input, train_target)
print(rf.oob_score_)
```

### 🐱 랜덤 포레스트 모델이 과적합되었어요.. 어떻게 해결할까요?
사용한 랜덤포레스트 모델이 과적합되었으니깐, 이제부터는 랜덤 포레스트의 변형들을 좀 알아보죠.
물론 그리드서치로 여러 파라미트를 조정하도 되지만요!
1. ExtraTreesClassifier
    - 각 결정 트리를 훈련할 때, 데이터를 무작위로 샘플링하고, 특성도 무작위로 선택합니다.
    - 랜덤 포레스트와 달리 기본적으로 부트스트랩 샘플링을 사용하지 않는다.
    - 일반적으로 특성의 무작위성을 추가하여 더 빠르게 훈련할 수 있는데, 훈련속도도 빠르지만 성능은 비슷하거나 더 나은 경우가 많다.
    - 모델이 학습한 후, 각 특성이 분류에 얼마나 기여했는지 평가하는 방법을 제공한다.

```python
from sklearn.ensemble import ExtraTreesClassifier
# 대부분 분류문제에서 max_feature를 sqrt로 지정하는데, 정수, 'log2'등으로 설정할 수 있다.
etc = ExtraTreesClassifier(n_jobs=-1, 
                           random_state=42,
                           max_features='sqrt')
scores = cross_validate(etc,
                        train_input,
                        train_target,
                        n_jobs=-1,
                        return_train_score=True)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
# 각 특성의 중요도
etc.fit(train_input, train_target)
etc.feature_importances_
```

```python
# 앞에서 설명했듯 그라디언트 부스팅은 깊이가 얕은 결정트리를 사용해 이전 트리의 오차를 보완하는 방식입니다.
# max_depth, learning_rate, n_estimators 파라미터의 변환을 통해 성능을 개선할 수 있습니다.
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier(random_state=42)
scores = cross_validate(gbc,
                        train_input,
                        train_target,
                        n_jobs=-1,
                        return_train_score=True)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
# 파라미터의 변환을 통한 성능 개선입니다. 결정 트리 개수를 5배 늘렸는데 과적합을 잘 억제하네요.
gbc = GradientBoostingClassifier(n_estimators=500,
                                 learning_rate=0.2,
                                 random_state=42)
scores = cross_validate(gbc,
                        train_input,
                        train_target,
                        n_jobs=-1,
                        return_train_score=True)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
# 그라디언트 부스팅이 결정트리나 랜덤포레스트보다 더 당도에 집중합니다.
gbc.fit(train_input, train_target)
gbc.feature_importances_
```

### 🐯 도대체 히스토그램 기반 그레이디언트 부스팅이 뭔가요..

    일반적을 히스토그램 기반 그레이디언트 부스팅은 기본 매개변수에서 안정적인 성능을 자랑하는데,
    max_iter는 부스팅 반복횟수를 의미하니 해당 매개변수만 바꿔서 사용해보세요.

```python
import matplotlib.pyplot as plt
x = np.random.randint(0, 10, size=(100))
plt.hist(x, bins=10)
```

```python
from sklearn.ensemble import HistGradientBoostingClassifier
# 기본 max_iter은 100
hgbc = HistGradientBoostingClassifier(max_iter=50,
                                      random_state=42)
scores = cross_validate(hgbc,
                        train_input,
                        train_target,
                        n_jobs=-1,
                        return_train_score=True)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

    - permutation_importance 기본 원리:
        모델이 훈련된 후, 각 특성의 중요도를 평가하기 위해 해당 특성을 무작위로 섞어서 모델의 성능 변화를 관찰합니다. 특성을 하나씩 랜덤하게 섞어서 모델의 성능의 변화를 관찰해 어떤 특성이 중요한지 반환합니다.
        원래의 데이터셋에서 각 특성이 얼마나 중요한지를 평가하기 위해, 
        특정 특성을 섞은 후 모델의 성능이 얼마나 감소하는지를 측정함
        성능이 크게 감소하면 해당 특성이 중요한 것으로 간주하고, 성능 변화가 적으면 덜 중요한 것으로 간주함
    - 모델 독립성:
        permutation_importance는 모델에 구애받지 않아서, 회귀, 분류, 앙상블 모델 등 
        다양한 모델에 사용할 수 있습니다.. 예를 들어, 다음과 같은 모델에서 사용할 수 있습니다.
        결정 트리 (Decision Tree)
        랜덤 포레스트 (Random Forest)
        그래디언트 부스팅 (Gradient Boosting)
        로지스틱 회귀 (Logistic Regression)
        서포트 벡터 머신 (SVM)
        신경망 (Neural Networks)

```python
# 특성 중요성
from sklearn.inspection import permutation_importance
hgbc.fit(train_input, train_target)
# n_repearts 각 특성에 대해 무작위로 섞는 과정의 횟수. 신뢰성을 높인다.
result = permutation_importance(hgbc,
                                train_input,
                                train_target,
                                n_jobs=-1,
                                n_repeats=10,
                                random_state=42)
print(result.importances_mean)
```

```python
# 테스트의 특성에 중요도입니다.
result = permutation_importance(hgbc,
                               test_input,
                               test_target,
                               n_jobs=-1,
                               n_repeats=10,
                               random_state=42)
print(result.importances_mean)
```

```python
# 결정 트리보다 1정도 점수가 올랐네요!
hgbc.score(test_input, test_target)
```

```python
# 지금까지는 사이킷런을 사용해서 다른 라이브러리를 사용해서 구현할 수 있습니다.
from xgboost import XGBClassifier
xgb = XGBClassifier(tree_method='hist',
                    random_state=42)
scores = cross_validate(xgb,
                        train_input,
                        train_target,
                        return_train_score=True)
print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```
